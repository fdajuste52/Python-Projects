## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
 SELECT COUNT(*) FROM CHICAGO_CRIME_DATA;
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
 SELECT * FROM CHICAGO_CRIME_DATA FETCH FIRST 10 ROWS ONLY;
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
 SELECT CASE_NUMBER, DESCRIPTION FROM CHICAGO_CRIME_DATA WHER DESCRIPTION LIKE '%MINOR%';
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
SELECT PRIMARY_TYPE, DESCRIPTION FROM CHICAGO_CRIME_DATA WHERE DESCRITPION LIKE '%CHILD%' AND PRIMARY_TYPE='KIDNAPPING';
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
SELECT DISTINCT(PRIMARY_TYPE) FROM CHICAGO_CRIME_DATA WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%';
```

### Problem 6

##### List the average safety score for each type of school.


```python
SELECT "ELEMENTARY, MIDDLE, OR HIGH SCHOOL", AVG(SAFETY_SCORE) AS AVERAGE_SAFETY FROM CHICAGO_PUBLIC_SCHOOLS GROUP BY "ELEMENTARY, MIDDLE, OR HIGH SCHOOL";
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY FROM CHICAGO_CENSUS_DATA ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5;
```

### Problem 8

##### Which community area is most crime prone?


```python
SELECT CCD.COOMUNITY_AREA_NUMBER, COUNT(CCD.COMMUNITY_AREA_NUMBER) AS CRIME_PRONE FROM CHICAGO_CRIME_DATA as CCD GROUP BY CCD.COMMUNITY_AREA_NUMBER ORDER BY COUNT(COMMUNITY_AREA_NUMBER) DESC LIMIT 1; 
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA WHERE COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA AS CCD GROUP BY CCD.COMMUNITY_AREA_NUMBER ORDER BY COUNT(CCD.COMMUNITY_AREA_NUMBER) DESC LIMIT 1) LIMIT 1;
```
